```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# how many observations we are going to use in ML. need to assign it. let say 500
# y = w1x1 + w2x2 + b + noise => (y = 5x1 - 4x2 + 11).. this is our model which we try to replicate with algorithm 

total_observation = 500
```

```python
# assign dependent variables (x1 and x2) the uniform values
x1 = np.random.uniform(-10,10,(total_observation,1))
x2 = np.random.uniform(-10,10,(total_observation,1))
```

```python
# and input variable is [x1 x2]

inputs = np.column_stack((x1,x2))
```

```python
# need to add some noise in model otherwise it will predict the model as such

noise = np.random.uniform(-1,1,(total_observation,1))
```

```python
# then need to define the target data. which come from y = 5x1 - 4x2 + 11

target = 5*x1 - 4*x2 +11 + noise

target.shape
```

```python
# 3d graph for x1,x2 and target
target = target.reshape(total_observation,)
fig = plt.figure()
axis = fig.add_subplot(111,projection='3d')
axis.plot(x1,x2,target)
axis.view_init(elev=300)
target = target.reshape(total_observation,1)
```

# Gradient Descent
# Train the model

```python
# manual gradient decent calculation 
# before , we need to decide a random value for bias and weight.
# as per the rule, dimension for bias and weight would be - 
# target = xw + b => m*n = (n*k)(k*m) + m
# where m = number of output variables
# k = number of input variables
# n = number of observations

# so in our scenario, dimensions of weight and bias are (k*m) and (m) which are (2,1) and (1) respectively

weight =np.random.uniform(-.1,.1,size=(2,1))
bias = np.random.uniform(-.1,.1,size=1)
print(weight)
print(bias)
```

```python
# now plan is to find weight and bias. calculate the y with random values of weight and bias.
# and compare y with target to find the loss. and accordingly adjust weight and bias 
# and do it in cycle till we get loss equivalent to zero.

# calculate output
# calculate loss
# adjust weight and bias wtih gradiaent descent concept
l_rate = 0.02

for i in range(200):
    y = np.dot(inputs,weight) + bias
    loss = y - target
    loss2 = np.sum(loss**2)/2/total_observation
    print(loss2)
    loss_average = loss / total_observation
    weight = weight - l_rate*np.dot(inputs.T,loss_average)
    bias = bias - l_rate* np.sum(loss_average)
```

```python
print(f'weight and bias are {weight} and {bias}')
```
